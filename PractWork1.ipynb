## Создание двух массивов случайных чисел на 30 элементов в диапазоне [-20,20]

```python
import numpy as NUMP
sizemas = 30
maxv=21
minv=-20
masX = NUMP.random.randint(minv,maxv, size=(sizemas))
masY =NUMP.random.randint(minv, maxv, size=(sizemas))
print "MasX: ", masX
print "MasY: ", masY
```

## Отрисовка графика по координатам из двух массивов

```python
import matplotlib.pyplot as mathpl

mathpl.figure()
mathpl.title('MasX and MasY')
mathpl.plot(masX,masY, 'green')
mathpl.show()
```

## Сортировка массива Х методом пузырька

```python
n = 1 
while n < len(masX):
    for i in range(len(masX)-n):
        if masX[i] > masX[i+1]:
            masX[i],masX[i+1] = masX[i+1],masX[i]
    n += 1
print "MasX: ", masX
```

## Отрисовка графика в точках по результатам сортировки

```python
mathpl.title('Bubble sorted masX, points')
mathpl.plot(masX, 'ro')
mathpl.show()
```
