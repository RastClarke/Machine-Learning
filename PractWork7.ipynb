## Кластеризация набора данных Mass-Shootings-Dataset.csv

```python
import graphlab
```

```python
data = graphlab.SFrame('Mass-Shootings-Dataset.csv')
```

```python
data.head()
```

## 1.Рассчитать TF-IDF для столбца Summary.

```python
data['word_count'] = graphlab.text_analytics.count_words(data['Summary'])
data.head()
```

```python
data['tf_idf']= graphlab.text_analytics.tf_idf(data['word_count'])
data
```

## 2. Заменить в столбце Mental Health Issues значения на [0, 1, 2], где:

###  0 - No
###  1 - Yes
###  2 - Unclear/Unknown и другие значения

```python
def transform_values(value):
    if value == 'No':
        return 0
    elif value == 'Yes':
        return 1
    else:
        return 2
```

```python
data['Mental Health Issues']=data['Mental Health Issues'].apply(transform_values)
```

```python
data
```

##  3. На основе примера обучить модель с использованием такого набора фич:
## [‘tf_idf’, ‘Total victims’, ’Mental Health Issues’] для

## -3 кластеров


```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(data['Total victims'], '.r')
```

```python
Myfeatures=['tf_idf', 'Total victims', 'Mental Health Issues']
model = graphlab.kmeans.create(data,
                               features=Myfeatures,
                               num_clusters=3)
```

```python
model['cluster_id'].head()
```

```python
graphlab.canvas.set_target('ipynb')
model.show()
```

```python
data['cluster'] = model['cluster_id']['cluster_id']
data
```

## -5 кластеров

```python
Myfeatures=['tf_idf', 'Total victims', 'Mental Health Issues']
modelfive = graphlab.kmeans.create(data,
                               features=Myfeatures,
                               num_clusters=5)
```

```python
modelfive['cluster_id'].head()
```

```python
graphlab.canvas.set_target('ipynb')
modelfive.show()
```

```python
data['clusterfive'] = modelfive['cluster_id']['cluster_id']
data
```
