```python
import graphlab
```

```python
products = graphlab.SFrame('Industrial_&_Scientific.csv')
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['text'])
```

```python
selected_words = ['bad', 'awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible' , 'terrible', 'awful', 'wow', 'hate']
products['word_count'].head()

def selected_words_counter(strok):
    count ={}
    
    for w in selected_words:
        count[w] = strok.get(w, 0)
    
    return count

products["selected_words_count"] = products["word_count"].apply(selected_words_counter)        
            
products          
```

2. Создать новую модель, используя в качестве фичи новый список слов из пункта 1.

```python
products = products[products['score']!=3]
products['sentiment'] = products['score']>=4

train_data,test_data = products.random_split(.8, seed=0)

sentiment_model = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=['selected_words_count'],
                                                     validation_set=test_data)
```

3. Сравнить метрики полученные в результате проверки первой и новой модели

Результаты новой модели

```python
sentiment_model.evaluate(test_data, metric='roc_curve')
```

```python
graphlab.canvas.set_target('ipynb')
sentiment_model.show(view='Evaluation')
```

Построение старой модели

```python
products = products[products['score']!=3]
products['sentiment'] = products['score']>=4

train_data,test_data = products.random_split(.8, seed=0)
sentiment_model_first = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=['word_count'],
                                                     validation_set=test_data)
```

Результаты старой модели

```python
sentiment_model_first.evaluate(test_data, metric='roc_curve')
```
